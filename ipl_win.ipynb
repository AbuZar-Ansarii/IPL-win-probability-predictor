```python
import pandas as pd
import numpy as np
```

```python
matche = pd.read_csv("matches.csv")
delivery = pd.read_csv("deliveries.csv")
```

```python
matche.head()
```

```python
matche.shape
```

```python
delivery.head()
```

```python
delivery.shape
```

```python
total_score_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
matche = matche.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
matche.head()
```

```python
matche["team1"].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
matche['team1'] = matche['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
matche['team2'] = matche['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
matche['team1'] = matche['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
matche['team2'] = matche['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
matche = matche[matche['team1'].isin(teams)]
matche = matche[matche['team2'].isin(teams)]
```

```python
matche.shape
```

```python
matche = matche[matche["dl_applied"] == 0]
```

```python
matche.shape
```

```python
matche.head()
```

```python
match_df = matche[['match_id','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
delivery_df.head()
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
delivery_df.shape
```

```python
delivery_df.head()
```

```python
delivery_df["current_score"] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
delivery_df['current_score'] = delivery_df['current_score'].astype(int) # Convert to numeric if necessary
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head(20)
```

```python
# Ensure 'player_dismissed' is of numeric type before groupby
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: 0 if x == '0' else 1).astype(int)

# Apply cumsum only to the 'player_dismissed' column within each group
delivery_df['wickets'] = 10 - delivery_df.groupby('match_id')['player_dismissed'].cumsum()

# Display the first 10 rows
delivery_df.head(10)
```

```python
delivery_df.head()
```

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df.head()
```

```python
def result(row):
  return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df.head()
```

```python
final_dlv_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
final_dlv_df.head()
```

```python
# shuffling
final_dlv_df = final_dlv_df.sample(final_dlv_df.shape[0])
```

```python
final_dlv_df.sample()
```

```python
final_dlv_df.dropna(inplace=True)
```

```python
final_dlv_df = final_dlv_df[final_dlv_df['balls_left'] != 0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = final_dlv_df.drop(columns=['result'])
y = final_dlv_df['result']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
